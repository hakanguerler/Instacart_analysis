```python
#When too many users make Instacart orders at the same time, the app freezes. The senior technical officer 
#at Instacart wants you to 
#identify the busiest hours of the day. Rather than by hour, they want periods of time labeled 
#“Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels 
#called “busiest_period_of_day.”
```

```python
df_ords_prods_merge.head()
```

```python
labeling_counts = df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
labeling_counts
```

```python
list_quantile = list(df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False).quantile([0.75, 0.5, 0.25]))
#using percentages
```

```python
list_quantile
```

```python
result_labeling_hour = []

for value in df_ords_prods_merge['order_hour_of_day']:
    for key, val in labeling_counts.items():
        if value == key:
            if val < list_quantile[2]:
                result_labeling_hour.append('Less orders')
            elif val >= list_quantile[2] and val < list_quantile[1]:
                result_labeling_hour.append('Average orders')
            elif val >= list_quantile[1]:
                result_labeling_hour.append('Most orders')
```

```python
print(result_labeling_hour)
```

```python
def labeling_result (x):
    most_orders = x.count('Most orders')
    average_orders = x.count('Average orders')
    less_orders = x.count('Less orders')
    
    print(f'Most orders: {most_orders}')
    print(f'Average orders: {average_orders}')
    print(f'Less orders: {less_orders}')
    print(f'Total: {most_orders + average_orders + less_orders}')
```

```python
labeling_result(result_labeling_hour)
```

```python
df_ords_prods_merge['order_hour_of_day'].shape
```

```python
df_ords_prods_merge['busiest_period_of_day'] = result_labeling_hour
```
